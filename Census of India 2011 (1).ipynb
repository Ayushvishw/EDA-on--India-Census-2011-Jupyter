### Q1) Which are the 10 most populated states of India?

```python
sum_population = final_data.groupby(["State name"])["Population"].sum().sort_values(ascending = False).head(10)

#Plotting bar using matplotlib
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(20,7))
colors = ['darkorange','darkorange','orange','whitesmoke','lightskyblue','lightskyblue','whitesmoke','green','darkgreen','darkgreen']
plt.plot(sum_population.index, sum_population, color = 'black', marker = "o")
plt.bar(sum_population.index, sum_population, color = colors, width = 0.6, edgecolor = "black", linewidth = 2)
plt.title('Top 10 most populated states', c='black', fontsize = 25)
plt.xlabel('States', c='black', fontsize = 12)
plt.ylabel('Population', c='black', fontsize = 12);

```

#### Observation 
* Uttar pradesh is the state with highest of 20.42 crore as per last census occured in 2011
* Gujrat is on the 10th number of top list with a population of a approx 6.5 crore

### Q2 Which are the 5 least populated states?

```python
least =  final_data.groupby(["State name"])["Population"].sum().sort_values(ascending = True).head()
```

```python
plt.figure(figsize=(20,7)) 
colors = ['darkorange','darkorange','whitesmoke','darkgreen','darkgreen']
plt.plot(least.index, least, color = 'slategrey', marker = "o")
plt.bar(least.index, least, color = colors, width = 0.4, edgecolor = "black", linewidth = 2)
plt.title('5 Least Populated States', c='black', fontsize = 25)
plt.xlabel('States', c='black', fontsize = 16)
plt.ylabel('Population', c='black', fontsize = 16);
```

* Among all the states and union territories in India, Lakshdweep is the least populated with less then 1 lakh population

### Q3 What is the Percentage of men and women in the country Population?

```python
final_data.head(2)
```

```python
total_pop = final_data.Population.sum()
male_pop = final_data.Male.sum()
female_pop = final_data.Female.sum()
mens = (male_pop/total_pop)*100
womens = (female_pop/total_pop)*100
print( 'Number of male in whole population is {:.2f}'.format(mens),'%')
print( 'Number of female in whole population is {:.2f}'.format(womens),'%')
```

### Q4 What is the weight of different religions in the country?

```python
religious = ['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']
Hindus = final_data.Hindus.sum()
Muslims = final_data.Muslims.sum()
Christians = final_data.Christians.sum()
Sikhs = final_data.Buddhists.sum()
Jains = final_data.Jains.sum()
# PLoting

all_religious = np.array([Hindus, Muslims, Christians, Sikhs, Jains])
myexplode = (0.1,0.2,0.3,0.5,0.6,)
mycolors = ["darkorange", "darkgreen", "green", "limegreen", "lime"]
plt.pie(all_religious, labels = all_religious ,colors = mycolors, explode = myexplode, shadow = True, startangle = 30)
plt.title("Weight Of Different Religious in India")
plt.legend(religious, loc = 10, fontsize = 6)



```

* We can see Highest population in the country is of Hindus and lowest is of Jains

### Q5 Which are the Top 10 states with highest literacy rate?

```python
final_data.head()
```

```python
literacy_rate = final_data.groupby(['State name'])[['Population']].sum()
ltr_rate_df = literacy_rate
literate = final_data.groupby(['State name'])[['Literate']].sum()

ltr_rate_df['literate'] = literate["Literate"]
#new total literacy for each state added

ltr_rate_df['literacy_rate'] = (ltr_rate_df.literate/ltr_rate_df.Population)*100
literacy_rate_df = ltr_rate_df.sort_values('literacy_rate', ascending = False).head(10)

#Let's plot it


colors = ['darkorange','darkorange','orange','whitesmoke','lightskyblue','lightskyblue','whitesmoke','green','darkgreen','darkgreen']
plt.figure(figsize=(25,9))
plt.bar(literacy_rate_df.index, literacy_rate_df.literacy_rate, color=colors, width = 0.5, edgecolor = "black", linewidth = 2)
plt.title('Top 10 states/Uts with highest literacy rate', c='black', fontsize = 30)
plt.xlabel('state/Uts_name', c='black', fontsize = 16)
plt.ylabel('Literacy rate', c='black', fontsize = 16);
```

 * Kerela has the highest literacy rate among all the states

### Q6 Is there any relation between literacy rate and population rate?

```python
literacy_df = ltr_rate_df.sort_values("literacy_rate", ascending = False)


# Ploting by sns
plt.figure(figsize=(12,5))
sns.scatterplot(x = literacy_df.Population, y = literacy_df.literacy_rate, color="darkorange", s=100)
plt.title('Comparison between literacy rate and population rate', c='darkgreen', fontsize = 20)
```

* The graph denotes that almost all the place where literacy rate is high there population rate is low

### Q9)What is the weight of male and female workers in the country?

```python
mens_emp = final_data["Male_Workers"].sum()
womens_emp = final_data["Female_Workers"].sum()
gender_emp = [mens_emp, womens_emp]
#ploting 
labels = ("Males", "Females")
colors = "darkorange", "darkgreen"
plt.figure('figsize',(14,4))
plt.barh(labels, gender_emp, color = colors, edgecolor = "skyblue", linewidth = 2)
plt.xlabel("Weight Of Workers By Gender", color = "blue", fontsize = 21)

```

* Numbers of Male workers is almost 50% higher than womens 

### Q10) Workers by category 

```python
final_data.columns
```

```python
mains = final_data.Main_Workers.sum()
marginal = final_data.Marginal_Workers.sum()
non = final_data.Non_Workers.sum()
cultivator = final_data.Cultivator_Workers.sum()
agriculture = final_data.Agricultural_Workers.sum()
house = final_data.Household_Workers.sum()
others = final_data.Other_Workers.sum()
#ploting by sns
labels = ['Mains', 'Marginal', 'Non', 'Cultivator', 'Agriculture', 'House', 'Others']
workers = [mains, marginal, non, cultivator, agriculture, house, others]
sns.barplot(x = labels, y = workers)
plt.title("Weight Of Different Workers", fontsize = '21')
plt.xlabel("Types Of Workers")
plt.ylabel("Counts")

```

* The number of Non workers are high in india
